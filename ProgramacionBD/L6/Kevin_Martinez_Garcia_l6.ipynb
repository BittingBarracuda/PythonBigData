# Dask

```python
%%time
import dask.dataframe as dd
```

```python
%%time
df_dask = dd.read_csv('./yellow_tripdata_2018-01.csv', assume_missing = True)
df_dask.head()
```

```python
%%time 
df_dask.describe()
```

```python
%%time
df_dask.describe().compute()
```

```python
%%time
df_dask.fare_amount.value_counts()
```

```python
%%time
df_dask.fare_amount.value_counts().compute()
```

En general se observa una mejora en los tiempos con DASK que es especialmente significativa a la hora de leer el DataFrame.

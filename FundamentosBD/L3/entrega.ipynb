Comenzamos importando las librerías que vamos a usar durante las actividades.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

A continuación, voy a representar en una misma figura, las funciones $sin(x)$, $cos(x)$, $e^{x}$ y $x^{2}$ en el intervalo $[-5, 5]$. Comenzamos definiendo un conjunto de 100 puntos en ese intervalo y calculando las funciones usando numpy.

```python
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x)
y4 = np.power(x, 2)
```

Vamos a dibujar estas funciones en una figura con 4 ejes en total. Estos ejes estarán organizados en 2 filas y 2 columnas. Para mejor comprensión del código, vamos a utilizar los métodos de instancia de matplotlib (es decir, la orientación a objetos que proporciona esta biblioteca).

```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(12, 8)
fig.suptitle("Funciones de x")
# Títulos de las gráficas
titles = ['$sin(x)$', '$cos(x)$', '$e^{x}$', '$x^{2}$']
# Color con el que representaremos cada gráfica
colors = ["red", "blue", "orange", "black"]
# Valores de las funciones
values = [y1, y2, y3, y4]

k = 0
for i in range(2):
    for j in range(2):
        ax[i, j].set_title(titles[k])
        ax[i, j].plot(x, values[k], color = colors[k])
        k += 1

plt.show()
```

Para terminar, vamos a hacer un plot en 3D de la función $sin(\sqrt{x^{2} + y^{2}})$.

```python
x = np.arange(-10, 10, 0.25)
y = np.arange(-10, 10, 0.25)
# Creamos una cuadrícula con todos los puntos posibles
X, Y = np.meshgrid(x, y)
# Evaluamos la función en todos los puntos
f = np.sin(np.sqrt(X**2 + Y**2))

# Procedemos a dibujarla como una superífice 3D
# Creamos una figura de 10x10 pulgadas
fig = plt.figure(figsize = (10, 10))
# Añadimos la función a plotear como título
fig.suptitle("$sin(\sqrt{x^{2} + y^{2}})$")
# Generamos el eje sobre el que dibujar la función
ax = plt.axes(projection = '3d')
# Dibujamos la superficie con el esquema de color "coolwarm"
surface = ax.plot_surface(X, Y, f, cmap = "coolwarm")
fig.colorbar(surface)
plt.show()
```

Los diagramas de Caja-Bigotes (boxplots o box and whiskers) 
son una presentación visual que describe varias características 
importantes, al mismo tiempo, tales como la dispersión y simetría.

Para su realización se representan los tres cuartiles y los 
valores mínimo y máximo de los datos, sobre un rectángulo, 
alineado horizontal o verticalmente. A continuación, vamos a usar 
numpy para generar algunas distribuciones y matplotlib para
hacer un diagrama de cajas.

```python
# Distribución normal
normal = np.random.normal(size = 1000)
# Distribución Exponencial
binomial = np.random.exponential(size = 1000)
# Distribución poisson
poisson = np.random.poisson(size = 1000)

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(10, 5)
ax[0].set_title("Distribución Normal")
ax[0].boxplot(normal)

ax[1].set_title("Distribución Binomial")
ax[1].boxplot(binomial)

ax[2].set_title("Distribución Poisson")
ax[2].boxplot(poisson)

plt.show()
```
